## 🟢 STEP 3 — Missing values (Task 01)

```python
df['age']=df['age'].fillna(df['age'].median())
```

## 📊 5️⃣ UNIVARIATE ANALYSIS (Single column)

```python
# Age Distribution
sns.histplot(df['age'],kde=True, bins=30)
plt.title("Age Distribution")
plt.show()
# বোঝা যাবে:

# normal distribution নাকি skewed
# এখানে:

# bins=30 → age কে 30টা bar এ ভাগ করছে
# histogram → actual frequency দেখাচ্ছে
# kde=True → সেই histogram এর উপর smooth curve দিচ্ছে
```

```python
df['age'].skew()
# Interpretation:
# ~0 → normal distribution

# 0 → right skewed ✔️

# < 0 → left skewed
```

```python
df['age'].mean()
```

```python
df['age'].median()
# 👉 যদি:

# mean > median → right skewed ✔️
# mean ≈ median → normal
```

```python
# 🎯 Survival count
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()
```

## 6️⃣ BIVARIATE ANALYSIS (2 columns)

```python
# 🎯 Gender vs Survival
# 👉 দেখা যাবে:

# female বেশি survive করেছে কি না
sns.countplot(x='survived', hue='sex', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
#🎯 Age vs Fare
sns.scatterplot(x='age', y='fare', data=df)
plt.title("Age vs Fare")
plt.show()
```

```python
#🎯 Class vs Fare
sns.boxplot(x='class', y='fare', data=df)
plt.title("Fare by Class")
plt.show()
# 👉 outlier + distribution বোঝা যাবে
```

## 🔥 7️⃣ MULTIVARIATE ANALYSIS

```python
# 🎯 Correlation Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True),annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Correlation	 Meaning
# +1 এর কাছে	strong positive
# -1 এর কাছে	strong negative
# 0 এর কাছে	no relation
```

```python
# 🎯 Pairplot (very powerful)
```

```python
sns.pairplot(df[['age','fare','pclass','survived']], hue='survived')
plt.show()
```
