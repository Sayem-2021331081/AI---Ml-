STRING OPERATIONS FOR TEXT CLEANING



```python
df = pd.DataFrame({
    'name': ['  Alice ', 'BOB', 'charlie!!', 'DaVid  ', 'Eve@123'],
    'gender': [' Male', 'FEMALE ', 'male', 'FeMale', ' MALE'],
    'email': ['alice@gmail.com', 'bob@@mail.com', 'charlie@gmail.com',
              'david@gmail.com ', 'eve#gmail.com']
})

print(df)
```

1. Strip spaces (leading/trailing)

```python
df['name'] = df['name'].str.strip()
df['gender'] = df['gender'].str.strip()
df['email'] = df['email'].str.strip()

print(df)
```

2. Lowercase normalization

```python
df['name'] = df['name'].str.lower()
df['gender'] = df['gender'].str.lower()

print(df)
```

3. Remove special characters (regex cleaning)

```python
df['name'] = df['name'].str.replace(r'[^a-zA-Z ]', '', regex=True)
df['email'] = df['email'].str.replace(r'[^a-zA-Z0-9@.]', '', regex=True)

print(df)
```

4. Title case formatting (optional)

```python
df['name'] = df['name'].str.title()

print(df)
```
