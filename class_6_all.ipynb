```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Rahim', 'Karim', 'Mim', 'Sadia', 'Nayeem'],
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male'],
    'Salary': [50000, 40000, 60000, 45000, 55000],
    'Age': [22, 25, 23, None, 26]
})

print(df)
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department').agg({
    'Salary':['mean','max'],
    'Age':'count'
    
})
```

```python
df.groupby(['Department', 'Gender'])['Salary'].mean()
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'Name': ['Rahim', 'Karim', 'Mim']
})

print(df1)
```

```python
df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'Salary': [40000, 50000, 60000]
})

print(df2)
```

```python
pd.merge(df1, df2, on='id', how='inner')
```

```python
pd.merge(df1, df2, on='id', how='left')
```

```python
pd.merge(df1, df2, on='id', how='right')
```

```python
pd.merge(df1, df2, on='id', how='outer')
```

```python
pd.merge(
    df1,
    df2,
    left_on='user_id',
    right_on='id',
    how='left'
)
```

```python
df1 = pd.DataFrame({
    'Name': ['Rahim', 'Karim']
})

df2 = pd.DataFrame({
    'Name': ['Mim', 'Sadia']
})

pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```
