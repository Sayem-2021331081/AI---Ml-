```python
name=input('Enter Your name:')
no_of_subjects=int(input('Eneter number of subjects (1-5): '))
print(f"Welcome {name} ! You have {no_of_subjects} subjects")
score=[]
i=0
sum=0
while 1:
  if i==no_of_subjects:
    break
  curr_score=(int(input(f"Enter score for subject {i+1}: ")))
  if curr_score>=0 and curr_score<=100:
    score.append(curr_score)
    i+=1
    sum+=curr_score

  else :
    print(f'Your score for subject {i+1} is out of range')
    print('Please retrasmit your score')
    continue
average=float((sum*1.0)/no_of_subjects)
ss=sorted(score)

def assign_grade(average):
  grade=''
  if average>=80:
    grade='A'
  elif average>=65:
    grade='B'
  elif average>=50:
    grade='C'
  elif average>=35:
    grade='D'
  else:
    grade='F'
  return grade

passed_sub=[score for i in range(len(score)) if score[i]>=50]

print(" ")

print('========== Grade Report ==========')

print(f'Student  : {name}')
print(f'Scores   : {score}')
print(f'Total    : {sum}')
print(f'Average  : {average:.2f}')
print(f"Highest  : {ss[len(score)-1]}  | Lowest : {ss[0]}")
print(f'Grade    : {assign_grade(average)}')
print(f'Passed   : {len(passed_sub)} out of {len(score)} subjects')






```

```python
n=int(input())
# a
inventory={}
for i in range(n):
  product_name=input()
  price=float(input())
  category=input()
  inventory[product_name]={"price":price,"category":category}
print('--- Available Products ---')
for product_name,info in inventory.items():
  print(f'{product_name}  :৳ {info["price"]:.2f} [{info["category"]}]')



# b
print('--- Available Categories ---')
categories=set()

for product,info in inventory.items():
  categories.add(info['category'])

print(categories)
# c
Cart=[]
while True:
  curr_product=input('Enter product ( done or finish):')
  if curr_product == 'done':
    break
  ch=0
  for product_name in inventory:
    if product_name==curr_product :
      ch=1
      break
  if not ch:
    print(f'⚠ {curr_product} not found in inventory. Try again.')
  else :
    Cart.append(curr_product)

print(Cart)
# d

def calculate_total(cart, inventory):
  total=0
  for item in cart:
        total += inventory[item]['price']
  return total

total=calculate_total(Cart,inventory)
print(f'Total (before discount): ৳ {total}')

# e
dis=0
if total>=5000:
  dis=0.15
elif total>=2000:
  dis=0.10
elif total>=1000:
  dis=0.05
dis_amount=total*dis
final_amount=total-dis_amount

if dis==0.15:
  print('Discount applied : 15% off')
elif dis==0.10:
  print('Discount applied : 10% off')
elif dis==0.05:
  print('Discount applied : 5% off')
else: print('Discount applied : 0% off')

print(f'Discount amount  : ৳  {dis_amount}')
print(f'Final amount     : ৳ {final_amount}')

# f
cart_summary={}

for item in Cart:
  cnt=0
  for i in Cart:
    if item==i:
      cnt+=1
  cart_summary[item]=cnt

print(f'Cart Summary : {cart_summary}')


# g
# final_dict={}
# for product,info in inventory.items():
#   final_dict[info['price']]=product

# sorted_dict = dict(sorted(final_dict.items(),reverse=True))
# items=()
# for item in inventory.values():
#   items+=(item,)

# print(items)
items=list(inventory.items())
items.sort(key=lambda x:x[1]['price'])
top3=()
cnt=0
for i in range(len(items)-1,-1,-1):
  top3+=(items[i][0],)
  cnt+=1
  if cnt==3:
    break

print(f'Top 3 most expensive items: {top3}')

















```
