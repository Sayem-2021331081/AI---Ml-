```python
scores=[72, 45, 88, 60, 91, 33, 78, 55]
c_scores=scores.copy()
c_scores.sort(key=lambda x: x, reverse=True)
print(f"Sorted(high -> low):{c_scores}")

```

```python
P_scores=[x for x in scores if x>=60]
print(f"Passing Scores (>=60): {P_scores}")
```

```python
U_scores=[x+5 for x in scores]
print(f"Scores after +5 bonus: {U_scores}")
```

```python
sum_scores=sum(scores)
print(f"Total of all Scores: {sum_scores}")
```

```python
grades=['A', 'B', 'A', 'C', 'B', 'A', 'D', 'C']
u_grades=set(grades)
u_grades=list(u_grades)
u_grades.sort()
print(f"Unique Grades: {u_grades}")

```

```python

import csv


s_records={"Alice":88,"Bob":72,"Charlie":45,"David":91,"Eve":78}
with open("marks.csv","w",newline="") as f:
    writer=csv.writer(f)
    writer.writerow(["Name","Scores"])
    for name,score in s_records.items():
        writer.writerow([name,score])

print("--- marks.csv ---")
with open('marks.csv','r') as f:
    reader =csv.reader(f)
    for row in reader:
        print(row)
```
