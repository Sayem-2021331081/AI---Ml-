```python
scores=[72, 45, 88, 60, 91, 33, 78, 55]
c_scores=scores.copy()
c_scores.sort(key=lambda x: x, reverse=True)
print(f"Sorted(high -> low):{c_scores}")

```

```python
P_scores=[x for x in scores if x>=60]
print(f"Passing Scores (>=60): {P_scores}")
```

```python
U_scores=[x+5 for x in scores]
print(f"Scores after +5 bonus: {U_scores}")
```

```python
sum_scores=sum(scores)
print(f"Total of all Scores: {sum_scores}")
```

```python
diary=dict()
while True:
    s=input()
    if not s:
        break
    date,m=s.split(':', 1)
    diary[date]=m

print("--- diary.txt ---")
for d,m in diary.items():
    print(f"{d}: {m}")
    
```

```python
try:
    with open("diary.txt",'w') as f:
      for d,m in diary.items():
        f.write(f"{d}: {m}\n")
except Exception as e:
    print(f"Error writing to file: {e}")

try:
    with open("diary.txt", "r") as f:
        print(f.read())

except FileNotFoundError:
    print("File not found")
```
