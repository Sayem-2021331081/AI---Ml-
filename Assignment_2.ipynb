```python
entries=[]
f_dates=[]
with open('diary.txt', 'r') as f:
    for line in f:
        entries.append(line.strip())
        f_dates.append(line.split()[0])
```

```python
print(f"total entries: {len(entries)}")
print(f"dates: {f_dates}")
```

```python
ls=""
mx=0
for s in entries:
    if len(s) > mx:
        mx = len(s)
        ls = s
print(f"longest entry: {ls}")
```

```python
import json
summary={
    "total_entries": len(entries),
    "dates": f_dates,
    "longest_entry": ls
}

with open('summary.json', 'w') as f:
    json.dump(summary, f, indent=1)

with open('summary.json', 'r') as f:
    data = json.load(f)

print(json.dumps(data, indent=1))    
```
